# Formas

* 1 - Método interno do Python: csv

```python
import csv

with open('advertising.csv') as arquivo:
    arquivo_csv = csv.reader(arquivo, delimiter = ",")
    for i, linha in enumerate(arquivo_csv):
        if i == 0:
            print("Cabeçalho: " + str(linha))
        else:
            print("Valor: " + str(linha))
```
